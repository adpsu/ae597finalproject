```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re

base_url = 'https://www.fueleconomy.gov/feg/PowerSearch.do?action=noform&year1=2011&year2=2024&cbmcsmallCars=Small+Cars&cbmcfamilySedans=Family+Sedans&cbmcupscaleSedans=Upscale+Sedans&cbmcluxurySedans=Luxury+Sedans&cbmclargeSedans=Large+Sedans&cbmchatchbacks=Hatchbacks&cbmcstationWagons=Station+Wagons&cbmcpickupTrucks=Pickup+Trucks&cbmcsportUtilityVehicles=Sport+Utility+Vehicles&cbmcminivans=Minivans&cbmcvans=Vans&minmsrpsel=0&maxmsrpsel=100000&city=0&hwy=0&comb=0&cbvtgasoline=Gasoline&cbvthybrid=Hybrid&cbvtplugin=Plug-in+Hybrid&cbvtelectric=Electric&YearSel=2011-2024&make=&mclass=Small+Cars%2C+Family+Sedans%2C+Upscale+Sedans%2C+Luxury+Sedans%2C+Large+Sedans%2C+Hatchbacks%2C+Station+Wagons%2C+Pickup+Trucks%2C+Sport+Utility+Vehicles%2C+Minivans%2C+Vans&vfuel=&vtype=Gasoline%2C+Hybrid%2C+Plug-in+Hybrid%2C+Electric&trany=&drive=&cyl=&MpgSel=000&sortBy=Comb&Units=&url=SearchServlet&opt=new&minmsrp=0&maxmsrp=100000&minmpg=0&maxmpg=0&sCharge=&tCharge=&startstop=&cylDeact=&rowLimit=200'

urls = []
for page_no in range(1, 62):  # From page 1 to 61
    if page_no == 1:
        # The first page has a slightly different URL structure
        url = base_url + '&action=PowerSearch'
    else:
        url = base_url + f'&pageno={page_no}&tabView=0'
    urls.append(url)
```

```python
data = []
x1 = []
vehicle_class = 'unknown'

for url in urls:
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    # Extract all vehicle entries
    vehicle_entries = soup.find_all('tr')
    

    for entry in vehicle_entries:
        car_image = entry.find('img', class_='img-thumbnail img-responsive veh-photo')
        range_cell = entry.find('td', class_='range')
        config_span = entry.find('span', class_='config')
        
        if config_span:
            config_text = config_span.get_text()
            # Split the text and extract the part you need, e.g., "Gasoline"
            parts = config_text.split(',')
            for part in parts:
                vehicle_class = 'Gasoline' if "Gasoline" in part else vehicle_class
                
        if car_image:
            car_name = car_image['alt']
            msrp_cell = entry.find('td', class_='msrp')
            msrp_text = msrp_cell.get_text(strip=True) if msrp_cell else 'Not Available'

            # Extract the model year and update the car name
            model_year_match = re.match(r'(\d{4})\s+(.*)', car_name)
            model_year = model_year_match.group(1) if model_year_match else None
            car_name = model_year_match.group(2) if model_year_match else car_name
            
            vehicle_class = 'EV' if entry.find('table', class_='ev') else vehicle_class

            # Extract MSRP value (If there is a range, then the maximum values are taken)
            min_msrp_value = None
            max_msrp_value = None
            msrp_matches = re.findall(r'\$\d+,\d+|\$\d+', msrp_text)
            if len(msrp_matches) > 1:
                min_msrp_value = int(msrp_matches[0].replace('$', '').replace(',', ''))
                max_msrp_value = int(msrp_matches[-1].replace('$', '').replace(',', ''))
            elif len(msrp_matches) == 1:
                min_msrp_value = int(msrp_matches[0].replace('$', '').replace(',', ''))
                max_msrp_value = min_msrp_value
                
                
        total_range = 'Not Available'
        if range_cell:
            vehicle_class = 'PHEV' if entry.find('td', class_='range phev') else vehicle_class
            if vehicle_class == 'EV':
                total_range_div = entry.find('div', class_='totalRange')
                if total_range_div:
                    x1.append(1)
                    total_range = total_range_div.get_text(strip=True).split()[0]
            elif vehicle_class == 'PHEV':

                cd_range_div = entry.find('div', class_='cdRange')
                if cd_range_div:
                    x1.append(2)
                    total_range = cd_range_div.get_text(strip=True).split()[0]
             

            data.append({
                'Model Year': model_year, 
                'Car Name': car_name, 
                'min_MSRP': min_msrp_value,
                'max_MSRP': max_msrp_value,
                'Vehicle Class': vehicle_class,
                'Total Range': total_range
            })

# Convert to DataFrame
df = pd.DataFrame(data)
df['MSRP'] = 0.5*(df['min_MSRP']+df['max_MSRP'])
#df = df.dropna(subset=['MSRP'])
df
```

```python
# Function to find duplicates, keep the row with maximum MSRP from each group, and include non-grouped data
def keep_max_msrp_include_others(dataframe):
    # Add a new column with the first three words of the car name
    dataframe['First Five Words'] = dataframe['Car Name'].apply(lambda x: ' '.join(x.split()[:5]))

    # Group by year and the first three words, and Total Range
    grouped = dataframe.groupby(['Model Year', 'First Five Words', 'Total Range'])

    # Initialize an empty DataFrame to store the results
    processed_rows = pd.DataFrame()

    for _, group in grouped:
        if len(group) > 2:
            # Keep the row with the maximum MSRP from each group
            max_msrp_row = group[group['MSRP'] == group['MSRP'].max()]
            processed_rows = pd.concat([processed_rows, max_msrp_row])
        else:
            # Include non-duplicate groups as is
            processed_rows = pd.concat([processed_rows, group])

    # Reset the index of the resulting DataFrame
    processed_rows.reset_index(drop=True, inplace=True)
    return processed_rows

# Apply the function to the DataFrame
df_combined = keep_max_msrp_include_others(df)

# Display the result
df_combined

```
